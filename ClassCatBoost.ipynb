```python
from catboost import CatBoostClassifier
```

```python
from google.colab import drive
import pandas as pd

# Google Drive'ı bağla
drive.mount('/content/drive')

# Dosyanın yolunu belirle
file_path = '/content/drive/My Drive/diabetes.csv'  # Dosyanın tam yolu

# Veriyi oku
df = pd.read_csv(file_path)
df = df.dropna()
df.head()
```

```python
y = df["Outcome"]
x = df.drop(["Outcome"], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)
```

```python
cat_model = CatBoostClassifier().fit(X_train, y_train)
```

```python
y_pred = cat_model.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
Cat_params = {"iterations": [200, 500, 100],
              "learning_rate": [0.01, 0.05, 0.1],
              "depth": [3,5,7]}
```

```python
Cat_CV_model = GridSearchCV(cat_model, Cat_params, cv=5, verbose=2).fit(X_train, y_train)
```

```python
Cat_CV_model.best_params_
```

```python
Cat_tuned = CatBoostClassifier(iterations = 100,
                               learning_rate = 0.1,
                               depth = 3).fit(X_train, y_train)
```

```python
y_pred_tunned = Cat_tuned.predict(X_test)
accuracy_score(y_test, y_pred_tunned)
```
